```python
import unicodedata

# Função ajustada para remover acentos e lidar com valores não-textuais
def remove_accents(input_str):
    # Convertendo o input para string para evitar erros com valores não-textuais
    input_str = str(input_str)
    nfkd_form = unicodedata.normalize('NFKD', input_str)
    return "".join([c for c in nfkd_form if not unicodedata.combining(c)])

# Supondo que 'df' seja o seu DataFrame
# Aplicando a função ajustada para transformar os valores de 'PERG.08'
df_pergs['PERG.8'] = df_pergs['PERG.8'].apply(remove_accents).str.upper()
```
